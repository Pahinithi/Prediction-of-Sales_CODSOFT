# Model Building

The model coefficients or model parameters, denoted as m values, represent the coefficients for the features in the equation of linear regression, where y is the response, c is the intercept, and m1 corresponds to the coefficient for the first feature, specifically TV in our case

In this case, we assign the feature variable, TV, to the variable X and the response variable, Sales, to the variable y as the initial step in model building using statsmodels.

```python
X = df['TV']
y = df['Sales']
```

Splitting the data into training and testing sets

 Assuming 'X' contains features and 'y' contains the target variable
 
 train_size = 0.7 specifies that 70% of the data will be used for training
 
test_size = 0.3 specifies that 30% of the data will be used for testing

 random_state = 100 ensures reproducibility of the split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

Displaying the first few rows of the training dataset

```python
X_train.head()
```

Assuming 'y_train' is  target variable (Sales) for the training set

```python
y_train.head()
```

# Building a Linear Model

```python
import statsmodels.api as sm
```

```python
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
```

Accessing the coefficients of the regression model

```python
lr.params
```

Displaying the summary of the linear regression model

```python
lr.summary()
```

# Summary

1. The coefficient for TV is 0.054, and its very low p-value indicates statistical significance, suggesting that the association is not merely coincidental.

2. The R-squared value is 0.816, signifying that 81.6% of the variance in Sales is accounted for by TV, which is considered a respectable R-squared value.

3. The F statistic's extremely low p-value implies that the model fit is statistically significant, affirming that the explained variance is not attributable solely to chance.

scatter plot of the training data

```python
# Scatter plot of the training data with red color
plt.scatter(X_train, y_train, color='red')

# Plotting the linear regression line with red color
plt.plot(X_train, 6.948 + 0.054 * X_train, 'r')

# Display the plot
plt.show()
```

# Assessing the model's performance is crucial for effective model evaluation.

Predicting the target variable on the training set

```python
y_train_pred = lr.predict(X_train_sm)
res = (y_train - y_train_pred)
```

Seaborn and Matplotlib to create a histogram

```python
# Create a new figure
fig = plt.figure()

# Use Seaborn to create a distribution plot of the residuals with red color
sns.distplot(res, bins=15, color='red')

# Set the title of the plot
fig.suptitle('Error Terms', fontsize=15)

# Set the x-axis label
plt.xlabel('y_train - y_train_pred', fontsize=15)

# Display the plot
plt.show()
```

```python
fig = plt.figure()

# Use Seaborn to create a distribution plot of the residuals with red color
sns.displot(res, bins=15, color='red')

fig.suptitle('Error Terms', fontsize=15)
plt.xlabel('y_train - y_train_pred', fontsize=15)

plt.show()
```

Scatter plot of the independent variable (X_train) against residuals

```python
# Scatter plot of the independent variable (X_train) against residuals with red points
plt.scatter(X_train, res, color='red')

# Set the title of the plot
plt.title('Residuals vs. Independent Variable', fontsize=15)

# Set the x-axis label
plt.xlabel('Independent Variable (X_train)', fontsize=12)

# Set the y-axis label
plt.ylabel('Residuals', fontsize=12)

# Display the plot
plt.show()
```

# Predictions for the Test Set

Augment X_test by adding a constant and then use the modified X_test to predict the corresponding y values with X_test_sm.

```python
X_test_sm = sm.add_constant(X_test)
y_pred = lr.predict(X_test_sm)
```

Display the first few predicted values

```python
y_pred.head()
```

Assuming 'y_test' contains the true target values for the test set

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

Printing the RMSE

```python
np.sqrt(mean_squared_error(y_test, y_pred))
```

Printing the R-squared value

```python
r_squared = r2_score(y_test, y_pred)
r_squared
```
